```python
# === After you have acc_veh_per from the merge ===
# acc_veh_per = pd.merge(df2, pd.merge(df3, df1, on="CASENUM", how="left"), on=["CASENUM","VEH_NO"], how="left")

import pandas as pd

def pick_first_present(df, candidates):
    """Return the first column name from candidates that exists in df, else None."""
    for c in candidates:
        if c in df.columns:
            return c
    return None

# --- Define desired features (by alias groups to avoid KeyErrors) ---
alias_groups = {
    # keys / target
    "target":           ["INJ_SEVNAME"],

    # accident (environment & context)
    "YEAR":             ["YEAR", "YEARNAME"],
    "MONTH":            ["MONTH", "MONTHNAME"],
    "DAY_WEEK":         ["DAY_WEEK", "DAY_WEEKNAME"],
    "HOUR":             ["HOUR", "HOURNAME"],
    "WEATHER":          ["WEATHERNAME", "WEATHER"],
    "LIGHT":            ["LGT_CONDNAME", "LGT_COND"],
    "RELJCT2":          ["RELJCT2NAME", "RELJCT2", "A_INTER", "A_INTSEC"],  # intersection context variants
    "TYP_INT":          ["TYP_INTNAME", "TYP_INT"],
    "REL_ROAD":         ["REL_ROADNAME", "REL_ROAD"],
    "WRK_ZONE":         ["WRK_ZONENAME", "WRK_ZONE"],
    "MAN_COLL":         ["MAN_COLLNAME", "MAN_COLL"],
    "REGION":           ["REGIONNAME", "REGION"],
    "URBANICITY":       ["URBANICITYNAME", "URBANICITY"],

    # vehicle (mechanics & roadway)
    "BODY_TYP":         ["BODY_TYPNAME", "BODY_TYP"],
    "ROLLOVER":         ["ROLLOVERNAME", "ROLLOVER"],
    "TRAV_SP":          ["TRAV_SPNAME", "TRAV_SP"],
    "VSPD_LIM":         ["VSPD_LIM", "VSPD_LIMNAME"],
    "VNUM_LAN":         ["VNUM_LAN", "VNUM_LANNAME"],
    "SPEEDREL":         ["SPEEDRELNAME", "SPEEDREL"],
    "VTRAFWAY":         ["VTRAFWAYNAME", "VTRAFWAY"],
    "VALIGN":           ["VALIGNNAME", "VALIGN"],
    "VPROFILE":         ["VPROFILENAME", "VPROFILE"],
    "VSURCOND":         ["VSURCONDNAME", "VSURCOND"],
    "VTRAFCON":         ["VTRAFCONNAME", "VTRAFCON"],

    # person/driver (demographics & protection)
    "AGE":              ["AGE"],
    "SEX":              ["SEXNAME", "SEX"],
    "PER_TYP":          ["PER_TYPNAME", "PER_TYP"],
    "REST_USE":         ["REST_USENAME", "REST_USE"],
    "HELM_USE":         ["HELM_USENAME", "HELM_USE"],
    "AIR_BAG":          ["AIR_BAGNAME", "AIR_BAG"],
    "EJECTION":         ["EJECTIONNAME", "EJECTION"],
    "DRINKING":         ["DRINKINGNAME", "DRINKING"],
    "DRUGS":            ["DRUGSNAME", "DRUGS"],

    # handy keys to keep
    "CASENUM":          ["CASENUM"],
    "VEH_NO":           ["VEH_NO"],
    "PER_NO":           ["PER_NO"],
}

# Resolve which actual columns exist
resolved_cols = []
for label, candidates in alias_groups.items():
    chosen = pick_first_present(acc_veh_per, candidates)
    if chosen:
        resolved_cols.append(chosen)

# Ensure target is present
if pick_first_present(acc_veh_per, alias_groups["target"]) is None:
    raise ValueError("Target column (e.g., 'INJ_SEVNAME') not found in merged data.")

# Build final list (prioritize ~30 core features + keys + target)
priority_order = [
    # target
    pick_first_present(acc_veh_per, alias_groups["target"]),

    # accident
    pick_first_present(acc_veh_per, alias_groups["YEAR"]),
    pick_first_present(acc_veh_per, alias_groups["MONTH"]),
    pick_first_present(acc_veh_per, alias_groups["DAY_WEEK"]),
    pick_first_present(acc_veh_per, alias_groups["HOUR"]),
    pick_first_present(acc_veh_per, alias_groups["WEATHER"]),
    pick_first_present(acc_veh_per, alias_groups["LIGHT"]),
    pick_first_present(acc_veh_per, alias_groups["RELJCT2"]),
    pick_first_present(acc_veh_per, alias_groups["TYP_INT"]),
    pick_first_present(acc_veh_per, alias_groups["REL_ROAD"]),
    pick_first_present(acc_veh_per, alias_groups["WRK_ZONE"]),
    pick_first_present(acc_veh_per, alias_groups["MAN_COLL"]),
    pick_first_present(acc_veh_per, alias_groups["REGION"]),
    pick_first_present(acc_veh_per, alias_groups["URBANICITY"]),

    # vehicle
    pick_first_present(acc_veh_per, alias_groups["BODY_TYP"]),
    pick_first_present(acc_veh_per, alias_groups["ROLLOVER"]),
    pick_first_present(acc_veh_per, alias_groups["TRAV_SP"]),
    pick_first_present(acc_veh_per, alias_groups["VSPD_LIM"]),
    pick_first_present(acc_veh_per, alias_groups["VNUM_LAN"]),
    pick_first_present(acc_veh_per, alias_groups["SPEEDREL"]),
    pick_first_present(acc_veh_per, alias_groups["VTRAFWAY"]),
    pick_first_present(acc_veh_per, alias_groups["VALIGN"]),
    pick_first_present(acc_veh_per, alias_groups["VPROFILE"]),
    pick_first_present(acc_veh_per, alias_groups["VSURCOND"]),
    pick_first_present(acc_veh_per, alias_groups["VTRAFCON"]),

    # person/driver
    pick_first_present(acc_veh_per, alias_groups["AGE"]),
    pick_first_present(acc_veh_per, alias_groups["SEX"]),
    pick_first_present(acc_veh_per, alias_groups["PER_TYP"]),
    pick_first_present(acc_veh_per, alias_groups["REST_USE"]),
    pick_first_present(acc_veh_per, alias_groups["HELM_USE"]),
    pick_first_present(acc_veh_per, alias_groups["AIR_BAG"]),
    pick_first_present(acc_veh_per, alias_groups["EJECTION"]),
    pick_first_present(acc_veh_per, alias_groups["DRINKING"]),
    pick_first_present(acc_veh_per, alias_groups["DRUGS"]),

    # keys
    pick_first_present(acc_veh_per, alias_groups["CASENUM"]),
    pick_first_present(acc_veh_per, alias_groups["VEH_NO"]),
    pick_first_present(acc_veh_per, alias_groups["PER_NO"]),
]

# Drop any None (in case something isn't present)
final_keep = [c for c in priority_order if c is not None]

# Subset to the final ~30 columns
df_model = acc_veh_per[final_keep].copy()

print("Kept columns ({}):".format(len(final_keep)))
print(final_keep)
print("Filtered shape:", df_model.shape)

# (Optional) quick cleanups for EDA:
import numpy as np
if "AGE" in df_model.columns:
    df_model["AGE"] = df_model["AGE"].replace({998: np.nan, 999: np.nan}).clip(lower=0, upper=100)

# (Optional) save
# df_model.to_csv("crss_eda_30cols.csv", index=False)
```

```python
from sklearn.utils import resample

# Drop tiny categories (<1237)
df_bal = df_model[df_model["INJ_SEVNAME"].isin([
    "No Apparent Injury (O)",
    "Possible Injury (C)",
    "Suspected Minor Injury (B)",
    "Suspected Serious Injury (A)",
    "Unknown/Not Reported",
    "Fatal Injury (K)"
])].copy()

# Target distribution before balancing
print("Before balancing:\n", df_bal["INJ_SEVNAME"].value_counts())

# Now downsample each class to 1237
balanced_dfs = []
for label, group in df_bal.groupby("INJ_SEVNAME"):
    balanced = resample(group, replace=False, n_samples=1237, random_state=42)
    balanced_dfs.append(balanced)

# Combine back
df_balanced = pd.concat(balanced_dfs, axis=0).reset_index(drop=True)

# Check final distribution
print("After balancing:\n", df_balanced["INJ_SEVNAME"].value_counts())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif

# ---------- 0) Start from your balanced frame ----------
df = df_balanced.copy()

# Target
target_col = "INJ_SEVNAME"
df = df.dropna(subset=[target_col]).reset_index(drop=True)

# Encode target
y = LabelEncoder().fit_transform(df[target_col].astype(str))

# Features
X = df.drop(columns=[target_col])

# ---------- 1) Split types ----------
is_categorical = X.dtypes.apply(lambda dt: (dt == 'object') or str(dt).startswith('category'))
cat_cols = X.columns[is_categorical].tolist()
num_cols = X.columns[~is_categorical].tolist()

# Optional: clean age-like numeric anomalies
if "AGE" in num_cols:
    X["AGE"] = X["AGE"].replace({998: np.nan, 999: np.nan}).clip(lower=0, upper=110)

# ---------- 2) Impute ----------
# Categorical: fill with "Missing" (works for object & category)
for c in cat_cols:
    if str(X[c].dtype).startswith("category"):
        # ensure "Missing" is an allowed category
        if "Missing" not in list(X[c].cat.categories):
            X[c] = X[c].cat.add_categories(["Missing"])
    X[c] = X[c].fillna("Missing").astype(str)

# Numeric: fill with median
for c in num_cols:
    if X[c].isna().any():
        X[c] = X[c].fillna(X[c].median())

# ---------- 3) Label-encode categoricals ----------
X_enc = X.copy()
for c in cat_cols:
    X_enc[c] = LabelEncoder().fit_transform(X_enc[c].astype(str))

# Safety: coerce any lingering weird types to numeric (shouldn't be needed, but safe)
X_enc = X_enc.apply(pd.to_numeric, errors="coerce")
for c in X_enc.columns:
    if X_enc[c].isna().any():
        # After encoding, any NaN left (rare) -> fill with column median (or 0)
        X_enc[c] = X_enc[c].fillna(X_enc[c].median())

# ---------- 4) Scores ----------
# Mutual Information: tell sklearn which columns were originally categorical
discrete_flags = [col in cat_cols for col in X_enc.columns]
mi_scores = mutual_info_classif(X_enc, y, discrete_features=discrete_flags, random_state=42)
mi_series = pd.Series(mi_scores, index=X_enc.columns)

# Pearson (linear correlation with encoded target)
pearson_series = X_enc.corrwith(pd.Series(y, index=X_enc.index))

# Combine and rank by MI
combined = pd.DataFrame({
    "Mutual_Info": mi_series,
    "Pearson": pearson_series
}).sort_values("Mutual_Info", ascending=False)

# ---------- 5) Show top features ----------
topN = 20
print("=== Top features by Mutual Information ===")
print(combined.head(topN).round(4))
```

```python
# Group by sex and severity, then count
severity_by_sex = (
    df_model.groupby(["SEXNAME", "INJ_SEVNAME"])
    .size()
    .reset_index(name="count")
)

# Pivot so each severity becomes a column
severity_table = severity_by_sex.pivot(
    index="SEXNAME", columns="INJ_SEVNAME", values="count"
).fillna(0)

print(severity_table)

# If you want proportions instead of raw counts:
severity_table_norm = severity_table.div(severity_table.sum(axis=1), axis=0)
print(severity_table_norm)
```
