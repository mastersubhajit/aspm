```python
motorcycle_body_types = list(range(80, 98))
```

```python
def combine_safety(row):
    """
    Create SAFETY_USENAME by combining restraint and helmet information.
    Rules:
      - If the vehicle is a motorcycle (BODY_TYP  motorcycle_body_types), 
        use helmet info (HELM_USENAME).
      - Else if REST_USENAME is 'None Used' or 'Not Applicable', 
        use HELM_USENAME instead.
      - Otherwise, use REST_USENAME.
    """
    rest = str(row.get('REST_USENAME', '')).strip()
    helm = str(row.get('HELM_USENAME', '')).strip()
    body = str(row.get('BODY_TYP', '')).strip()
    
    if body in motorcycle_body_types:
        # Motorcyclist  use helmet info
        return helm if helm else rest
    elif rest in ['None Used', 'Not Applicable', '', 'Unknown']:
        # Non-motorcyclist with no restraint  fallback to helmet info
        return helm if helm else rest
    else:
        # Default to restraint info
        return rest

# Apply to dataset
final_df['SAFETY_USENAME'] = final_df.apply(combine_safety, axis=1)

# Map to consistent human-readable categories
safety_map = {
    'DOT-Compliant Motorcycle Helmet': 'Helmet Used',
    'Helmet, Other Than DOT-Compliant Motorcycle Helmet': 'Helmet Used (Other)',
    'No Helmet': 'No Helmet',
    'Shoulder and Lap Belt Used': 'Seat Belt Used',
    'Shoulder Belt Only Used': 'Seat Belt Used',
    'Lap Belt Only Used': 'Seat Belt Used',
    'Child Restraint System - Forward Facing': 'Child Restraint',
    'Child Restraint System - Rear Facing': 'Child Restraint',
    'Booster Seat': 'Child Restraint',
    'None Used': 'No Restraint',
    'Restraint Used - Type Unknown': 'Unknown Restraint',
    'Not Applicable': 'No Restraint',
    'Unknown': 'Unknown Restraint'
}
final_df['SAFETY_USENAME'] = final_df['SAFETY_USENAME'].replace(safety_map)

# Final simplification to 'Protective Used' vs 'None'
final_df['SAFETY_USENAME'] = final_df['SAFETY_USENAME'].replace({
    'Helmet Used': 'Protective Used',
    'Helmet Used (Other)': 'Protective Used',
    'Seat Belt Used': 'Protective Used',
    'Child Restraint': 'Protective Used',
    'No Helmet': 'None',
    'No Restraint': 'None',
    'Unknown Restraint': 'None'
})

# Verify result
print(final_df['SAFETY_USENAME'].value_counts(dropna=False))
```

```python
import re

def _norm(x):
    if pd.isna(x):
        return ""
    return str(x).strip().lower()

# Mapping function
def map_impact_to_orientation(val):
    """
    Map IMPACT1NAME (which may be numeric codes, clock-point strings, or side codes)
    to one of: Front, Rear, LeftSide, RightSide, Top, Undercarriage, Other, Unknown.
    """
    if pd.isna(val):
        return "Unknown"

    # If it's numeric (int/float) use integer mapping
    try:
        if isinstance(val, (int, np.integer)):
            n = int(val)
            return _map_numeric_impact(n)
        # some values come as numeric strings
        if re.fullmatch(r"\d+", str(val).strip()):
            n = int(str(val).strip())
            return _map_numeric_impact(n)
    except Exception:
        pass

    s = _norm(val)

    # handle common textual patterns: "12 clock point", "12 o'clock", "12 clock"
    m = re.search(r"(\b\d{1,2}\b)", s)
    if m:
        n = int(m.group(1))
        return _map_numeric_impact(n)

    # detect explicit codes words
    if "top" in s:
        return "Top"
    if "undercarriage" in s or "undercarriage" in s:
        return "Undercarriage"
    if "left-front" in s or "left front" in s or "left-front side" in s:
        return "LeftSide"
    if "left-back" in s or "left back" in s:
        return "LeftSide"
    if "left" in s and "left-front" not in s and "left-back" not in s:
        return "LeftSide"
    if "right-front" in s or "right front" in s or "right-front side" in s:
        return "RightSide"
    if "right-back" in s or "right back" in s:
        return "RightSide"
    if "right" in s and "right-front" not in s and "right-back" not in s:
        return "RightSide"

    # other keywords
    if "non-collision" in s or "noncollision" in s:
        return "NonCollision"
    if "other" in s:
        return "Other"
    if "unknown" in s or "not reported" in s or s == "":
        return "Unknown"

    # fallback
    return "Other"

def _map_numeric_impact(n):
    """
    Numeric code mapping per CRSS:
    - 1..12 = clock points (1=1 o'clock etc)
    - 13 = Top, 14 = Undercarriage
    - 61 = Left, 62 = Left-Front Side, 63 = Left-Back Side
    - 81 = Right, 82 = Right-Front Side, 83 = Right-Back Side
    - other codes -> Other/Unknown
    """
    # clock points -> map to orientation:
    # clockface mapping: define groups:
    # front: 11,12,1
    # right: 2,3,4
    # rear: 5,6,7
    # left: 8,9,10
    if 1 <= n <= 12:
        if n in (11, 12, 1):
            return "Front"
        elif n in (2, 3, 4):
            return "RightSide"
        elif n in (5, 6, 7):
            return "Rear"
        elif n in (8, 9, 10):
            return "LeftSide"
    if n == 13:
        return "Top"
    if n == 14:
        return "Undercarriage"
    if n in (61,):
        return "LeftSide"
    if n in (62, 63):
        return "LeftSide"
    if n in (81,):
        return "RightSide"
    if n in (82, 83):
        return "RightSide"
    # codes like 18, 19 are object set-in-motion  treat as Other
    if n in (18, 19, 20, 55):
        return "Other"
    if n in (98, 99):
        return "Unknown"
    return "Other"
final_df['IMPACT_TYPE'] = final_df['IMPACT1NAME'].apply(map_impact_to_orientation)
final_df['IMPACT_TYPE'].value_counts()
```

```python
import re
def extract_speed(val):
    match = re.findall(r'\d+', str(val))
    return int(match[0]) if match else np.nan

final_df['SPEED_LIMIT'] = final_df['VSPD_LIMNAME'].apply(extract_speed)
final_df['SPEED_RATIO'] = final_df['TRAV_SP'] / final_df['SPEED_LIMIT']
final_df['SPEED_RATIO'].value_counts()
```

```python
final_df['DRINKINGNAME'] = (
    final_df['DRINKINGNAME']
    .astype(str)
    .str.strip()
    .replace({'nan':'Unknown','NaN':'Unknown','None':'Unknown','':''})
)

# Extract numeric speed limit if not already numeric
def extract_limit(val):
    if pd.isna(val): return np.nan
    m = re.findall(r"\d+", str(val))
    return float(m[0]) if m else np.nan

if 'SPEED_LIMIT' not in final_df.columns and 'VSPD_LIMNAME' in final_df.columns:
    final_df['SPEED_LIMIT'] = final_df['VSPD_LIMNAME'].apply(extract_limit)

# Compute ratio safely
final_df['SPEED_LIMIT'] = pd.to_numeric(final_df['SPEED_LIMIT'], errors='coerce')
final_df['TRAV_SP'] = pd.to_numeric(final_df['TRAV_SP'], errors='coerce')

final_df['SPEED_RATIO'] = (
    final_df['TRAV_SP'] / final_df['SPEED_LIMIT']
).replace([np.inf, -np.inf], np.nan)

# Fill missing/zero limits gracefully
final_df['SPEED_RATIO'] = final_df['SPEED_RATIO'].fillna(1).clip(0, 3)


bins = [0, 1.0, 1.2, 1.5, 2.0, np.inf]
labels = ['Normal', 'Mild', 'High', 'VeryHigh', 'Extreme']
final_df['SPEED_LEVEL'] = pd.cut(final_df['SPEED_RATIO'], bins=bins, labels=labels, include_lowest=True)

drink_map = {
    'Yes': 'Drinking',
    'No': 'Sober',
    'Suspected': 'Suspected',
    'Unknown': 'Unknown',
    'Refused': 'Refused',
    'Not Reported': 'Unknown',
}
final_df['DRINK_STATUS'] = final_df['DRINKINGNAME'].map(drink_map).fillna('Unknown')


final_df['ALCOHOL_SPEED_RISK'] = final_df['DRINK_STATUS'] + '_' + final_df['SPEED_LEVEL'].astype(str)

# Clean up any trailing _nan
final_df['ALCOHOL_SPEED_RISK'] = (
    final_df['ALCOHOL_SPEED_RISK']
    .str.replace('_nan','_Normal', regex=False)
    .str.replace('Unknown_nan','Unknown_Normal', regex=False)
)


collapse = {
    'Drinking_Extreme': 'High_Risk',
    'Drinking_VeryHigh': 'High_Risk',
    'Drinking_High': 'High_Risk',
    'Suspected_High': 'High_Risk',
    'Sober_Extreme': 'Speeding',
    'Sober_VeryHigh': 'Speeding',
    'Sober_High': 'Speeding',
}
final_df['ALCOHOL_SPEED_RISK'] = final_df['ALCOHOL_SPEED_RISK'].replace(collapse)

# Inspect
final_df[['DRINKINGNAME','SPEED_RATIO','SPEED_LEVEL','ALCOHOL_SPEED_RISK']]
```

| LGT_CONDNAME  | Records the type/level of light that existed at the time of the crash.                                                                                                                                      |
| WEATHERNAME   | Records the prevailing atmospheric conditions that existed at the time of the crash as indicated in the police crash report.                                                                                 |
| MAX_SEVNAME   | Records the single most severe injury of all people involved in the crash.                                                                                                                                   |

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

models = {
    "Decision Tree": DecisionTreeClassifier(
        max_depth=15,
        min_samples_split=2,
        min_samples_leaf=1,
        class_weight="balanced",
        random_state=42
    ),
    
    "Random Forest": RandomForestClassifier(
        n_estimators=400,
        max_depth=20,
        min_samples_split=2,
        min_samples_leaf=1,
        class_weight='balanced_subsample',
        n_jobs=-1,
        random_state=42
    ),
    
    "XGBoost": XGBClassifier(
        n_estimators=800,
        max_depth=10,
        learning_rate=0.05,
        objective='multi:softprob',
        eval_metric='mlogloss',
        enable_categorical=True,
        random_state=42
    ),
    
    "LightGBM": LGBMClassifier(
        n_estimators=400,
        num_leaves=80,
        learning_rate=0.08,
        max_depth=25,
        class_weight='balanced',
        force_row_wise=True,
        random_state=42
    ),
    
    "CatBoost": CatBoostClassifier(
        iterations=1000,
        depth=10,
        learning_rate=0.05,
        eval_metric='TotalF1',
        auto_class_weights='Balanced', 
        random_seed=42,
        verbose=200
    )
}

for name, model in models.items():
    print(f"\n{'='*70}")
    print(f"Training {name}...")
    
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    
    acc = accuracy_score(y_test, y_pred)
    print(f"\nFinal Accuracy ({name}): {acc*100:.2f}%")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=list(severity_map.keys())))
```

<h1>Randomized Search CV<h1>

```python
categorical_cols = [
    'STRATUMNAME', 'DEFORMEDNAME', 'SAFETY_USENAME',
    'ROLLOVERNAME', 'BODY_TYP', 'IMPACT1NAME', 'HOURNAME',
    'LGT_CONDNAME', 'VSPD_LIMNAME', 'REL_ROADNAME',
    'DRINKINGNAME', 'REGIONNAME'
]

for col in categorical_cols:
    if col in X_train.columns:
        X_train[col] = X_train[col].astype('category')
        X_test[col] = X_test[col].astype('category')
        print(f"Converted {col} to 'category' dtype.")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Define the parameter distributions for random search
param_distributions = {
    'n_estimators': [200, 400, 600],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    'class_weight': ['balanced', 'balanced_subsample']
}

# Initialize the model
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_distributions, 
    n_iter=25,                                # number of random combinations to try
    cv=3,                                     # 3-fold cross-validation
    scoring='accuracy',                       # or 'f1_macro' if imbalanced
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit the randomized search
random_search.fit(X_train_scaled, y_train)

# Print best parameters and score
print("Best Parameters:", random_search.best_params_)
print(f"Best Cross-Validation Accuracy: {random_search.best_score_:.4f}")

# Retrieve the best estimator
best_rf = random_search.best_estimator_

# Evaluate on test set
y_pred_best = best_rf.predict(X_test_scaled)
test_acc = accuracy_score(y_test, y_pred_best)
print(f"\nTest Accuracy: {test_acc*100:.2f}%")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best, target_names=list(severity_map.keys())))
```
